```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
counter = 0
rs,cs = fn.shape

fn.drop_duplicates(inplace=True)

if fn.shape==(rs,cs):
    print('\n\033[1mInference:\033[0m The dataset doesn\'t have any duplicates')
else:
    print(f'\n\033[1mInference:\033[0m Number of duplicates dropped/fixed ---> {rs-fn.shape[0]}')
```

```python
from sklearn.preprocessing import LabelEncoder
categ = ["mainroad","guestroom","basement","hotwaterheating","airconditioning","prefarea","furnishingstatus"]

# Encode Categorical Columns
le = LabelEncoder()
fn[categ] = fn[categ].apply(le.fit_transform)
```

```python
fn.head()
```

```python
corr = fn.corr()
plt.figure(figsize=(12,7))
sns.heatmap(corr,cmap='twilight_shifted',annot=True)
```

```python
X = fn.drop(['price'],axis=1)
y = fn['price']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
len(X_train),len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
score = r2_score(y_test,y_predict)
puff = mean_absolute_error(y_test,y_predict)
```
